```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
X = df['feature_0']
y = df['timestamp']
degree = 14

coeff = np.polyfit(X,y, deg = degree)
line = np.poly1d(coeff)
x_fit = np.linspace(X.min(), X.max(), 50)
y_fit = line(x_fit)

plt.scatter(X, y, label='Feature 0')
plt.plot(x_fit, y_fit, 'r', label=f'Fitted Polynomial (Degree {degree})')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Fitting')
plt.show()
```
