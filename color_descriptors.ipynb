```python
import cv2
import os
count=0
# Path to the directory containing your flower images
image_directory = '/content/drive/MyDrive/30820231'

# List of image filenames in the directory
image_files = os.listdir(image_directory)

# Initialize a dictionary to store filename-to-color-mapping
image_color_mapping = {}

# Loop through each image filename
for image_filename in sorted(image_files):
    # Check if the filename is in the expected format (number + '.jpg')
    if image_filename.endswith('.jpg.jpg'):
        # Read the image
        image_path = os.path.join(image_directory, image_filename)
        image = cv2.imread(image_path)

        # Check if the image was read successfully
        if image is not None:
            # Calculate mean color values
            mean_color = cv2.mean(image)
            image_color_mapping[image_filename] = {
                'mean_red': mean_color[2],
                'mean_green': mean_color[1],
                'mean_blue': mean_color[0]
            }
        else:
            print("Error reading image:", image_path)

# Print the mapping of filenames to mean color values
for filename, color_info in image_color_mapping.items():
    count+=1
    print(f"Image: {filename}, Mean Red: {color_info['mean_red']:.2f}, Mean Green: {color_info['mean_green']:.2f}, Mean Blue: {color_info['mean_blue']:.2f}")
print(count)
```
